```python
#genre lister
def genre_lister(object):
    import ast
    object=ast.literal_eval(object) #list is stored as string, converting String Representation of a List
    list_genre=[]
    for i in object:
        list_genre.append(i['name'])
    return list_genre
        
```

```python
movie['genres']=movie['genres'].apply(genre_lister)
```

```python
movie.head()
```

```python
movie['keywords']=movie['keywords'].apply(genre_lister)
```

```python
movie.head()
```

```python
#top 3 actors
def cc_lister(object):
    import ast
    object=ast.literal_eval(object) #list is stored as string, converting String Representation of a List
    list_cc=[]
    for i in object:
        list_cc.append(i['name'])
    return list_cc[:3] #taking top 3
        
```

```python
print(cc_lister(movie.loc[0,'cast']))
```

```python
movie['cast']=movie['cast'].apply(cc_lister)
```

```python
def director(object):
    import ast
    object=ast.literal_eval(object) #list is stored as string, converting String Representation of a List
    director=[]
    for i in object:
        if i['job'] == "Director":
            director.append(i['name'])
    return director
```

```python
print(director(movie.loc[0,'crew']))
```

```python
movie['crew']=movie['crew'].apply(director)
```

```python
movie.head()
```

```python
def stringcleaner(string):
    if isinstance(string, str): #checking if string
        string=string.replace(","," ")
        string=string.replace("?"," ")
        string=string.replace("."," ")
        string=string.replace("!"," ")
        return string.split()
    else:
        return []
```

```python
print(stringcleaner(movie.loc[0,'overview']))
```

```python
movie['overview']=movie['overview'].apply(stringcleaner)
```

```python
movie.head()
```

```python
movie['genres']=movie['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movie['keywords']=movie['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movie['cast']=movie['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movie['crew']=movie['crew'].apply(lambda x : [i.replace(" ","") for i in x])


```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()   #going to stem the text for word variations (actor, actors etc)
```

```python
#function to stem 
def stemmer(string):
    res=[]
    n=string.split()
    for i in n:
        res.append(ps.stem(i))
    
    return " ".join(res)
        
```

```python
stemmer(movie_df['tags'].loc[6])
```

```python
movie_df['tags']= movie_df['tags'].apply(stemmer)
```

```python
#test vectorization 
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english") 
#fitting and converting to numpyarray
vectors = cv.fit_transform(movie_df['tags']).toarray()
```

```python
cv.get_feature_names_out()
```

```python
#cosine distance between vectors  - the lesser the angle theta more similar the vectors -- cosine similarity (0 no 1 same)/ cosine distance 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity =  cosine_similarity(vectors)    #similarity matrix
```

```python
#for the first movie
similarity[0]    #diagonal 1 as movie getting compared to itself
```

```python
def recommend(movie):
    #fetching index of movie
    movie_index=movie_df[movie_df["title"]==movie].index[0] #index at 0th col
    distances=similarity[movie_index]
    #sorting using index+similarity from each element
    movielist=sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]  #top 5 most similar
    for i in movielist:
        print(movie_df.loc[i[0]].title) #title for each index
    
```

```python
recommend("Scream")
```
